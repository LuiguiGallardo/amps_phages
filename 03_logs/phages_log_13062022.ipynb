### Objetivos:
1.  Conocer cuáles AMPs están presentes en fagoma intestinal infantil
2.  Determinar AMPs diferenciales entre grupos NW, O y OMS

```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
# Coverage distribution
merged_cov_01 = pd.read_csv("../01_tables/merged_bwa_sorted_cov-0.1.txt", sep="\t")

coverage_distribution_table = pd.cut(merged_cov_01["coverage"], bins=18).value_counts()
print(coverage_distribution_table)

coverage_distribution = sns.displot(merged_cov_01, x=merged_cov_01["coverage"], bins=9)
coverage_distribution.set(title="Phages coverage", xlabel="Coverage (%)", ylabel="Number of phages")
plt.savefig("../02_graphs/coverage_distribution.svg")
plt.show()
plt.close()
```

```python
# Phages per sample
data_phages = pd.read_csv("../01_tables/data_sample_phages.txt", sep="\t")

colors = {"H": "c", "O": "y", "OMC": "r"}

# Number of phages
phages_sample = sns.boxplot(x=data_phages["group"], y=data_phages["phages"], palette=colors)
phages_sample = sns.swarmplot(x=data_phages["group"], y=data_phages["phages"], color=".25")
phages_sample.set(title="Number of phages mapped", xlabel="Group", ylabel="Number of phages")
plt.savefig("../02_graphs/phages_sample.svg")
plt.show()
plt.close()

# Absolute abundance
phages_abundance_sample = sns.boxplot(x=data_phages["group"], y=data_phages["absolute_abundance"], palette=colors)
phages_abundance_sample = sns.swarmplot(x=data_phages["group"], y=data_phages["absolute_abundance"], color=".25")
phages_abundance_sample.set(title="Absolute abundance of phages mapped", xlabel="Group", ylabel="Number of reads")
plt.savefig("../02_graphs/phages_abundance_sample.svg")
plt.show()
plt.close()

# Relative abundance 75
phages_abundance_sample = sns.boxplot(x=data_phages["group"], y=data_phages["relative_abundance_75"], palette=colors)
phages_abundance_sample = sns.swarmplot(x=data_phages["group"], y=data_phages["relative_abundance_75"], color=".25")
phages_abundance_sample.set(title="Relative abundance of phages mapped", xlabel="Group", ylabel="Relative abundance")
plt.savefig("../02_graphs/phages_relab_75_sample.svg")
plt.show()
plt.close()

# RPKM 75
phages_abundance_sample = sns.boxplot(x=data_phages["group"], y=data_phages["rpkm_75"], palette=colors)
phages_abundance_sample = sns.swarmplot(x=data_phages["group"], y=data_phages["rpkm_75"], color=".25")
phages_abundance_sample.set(title="RPKMs of phages mapped", xlabel="Group", ylabel="RPKMs")
plt.savefig("../02_graphs/phages_rpkm_75_sample.svg")
plt.show()
plt.close()

# Relative abundance 70
phages_abundance_sample = sns.boxplot(x=data_phages["group"], y=data_phages["relative_abundance_70"], palette=colors)
phages_abundance_sample = sns.swarmplot(x=data_phages["group"], y=data_phages["relative_abundance_70"], color=".25")
phages_abundance_sample.set(title="Relative abundance of phages mapped", xlabel="Group", ylabel="Relative abundance")
plt.savefig("../02_graphs/phages_relab_70_sample.svg")
plt.show()
plt.close()

# RPKM 70 
phages_abundance_sample = sns.boxplot(x=data_phages["group"], y=data_phages["rpkm_70"], palette=colors)
phages_abundance_sample = sns.swarmplot(x=data_phages["group"], y=data_phages["rpkm_70"], color=".25")
phages_abundance_sample.set(title="RPKMs of phages mapped", xlabel="Group", ylabel="RPKMs")
plt.savefig("../02_graphs/phages_rpkm_70_sample.svg")
plt.show()
plt.close()
```

```python
# Barplots
# Number of phages
phages_sample = sns.barplot(x=data_phages["sample"], y=data_phages["phages"], hue=data_phages["group"], palette=colors, dodge=False)
phages_sample.set(title="Number of phages mapped", xlabel="Group", ylabel="Number of phages")
plt.xticks(rotation='90', rotation_mode='anchor', ha='right')
plt.savefig("../02_graphs/phages_sample_barplot.svg")
plt.show()
plt.close()

# Absolute abundance
phages_abundance_sample = sns.barplot(x=data_phages["sample"], y=data_phages["absolute_abundance"], hue=data_phages["group"], palette=colors, dodge=False)
phages_abundance_sample.set(title="Absolute abundance of phages mapped", xlabel="Group", ylabel="Number of reads")
plt.xticks(rotation='90', rotation_mode='anchor', ha='right')
plt.savefig("../02_graphs/phages_abundance_sample_barplot.svg")
plt.show()
plt.close()

# Relative abundance 75
phages_abundance_sample = sns.barplot(x=data_phages["sample"], y=data_phages["relative_abundance_75"], hue=data_phages["group"], palette=colors, dodge=False)
phages_abundance_sample.set(title="Relative abundance of phages mapped", xlabel="Group", ylabel="Relative abudance")
plt.xticks(rotation='90', rotation_mode='anchor', ha='right')
plt.savefig("../02_graphs/phages_relab_75_sample_barplot.svg")
plt.show()
plt.close()

# RPKM 75
phages_abundance_sample = sns.barplot(x=data_phages["sample"], y=data_phages["rpkm_75"], hue=data_phages["group"], palette=colors, dodge=False)
phages_abundance_sample.set(title="RPKMs of phages mapped", xlabel="Group", ylabel="RPKMs")
plt.xticks(rotation='90', rotation_mode='anchor', ha='right')
plt.savefig("../02_graphs/phages_rpkm_75_sample_barplot.svg")
plt.show()
plt.close()

# Relative abundance 70
phages_abundance_sample = sns.barplot(x=data_phages["sample"], y=data_phages["relative_abundance_70"], hue=data_phages["group"], palette=colors, dodge=False)
phages_abundance_sample.set(title="Relative abundance of phages mapped", xlabel="Group", ylabel="Relative abudance")
plt.xticks(rotation='90', rotation_mode='anchor', ha='right')
plt.savefig("../02_graphs/phages_relab_70_sample_barplot.svg")
plt.show()
plt.close()

# RPKM 70
phages_abundance_sample = sns.barplot(x=data_phages["sample"], y=data_phages["rpkm_70"], hue=data_phages["group"], palette=colors, dodge=False)
phages_abundance_sample.set(title="RPKMs of phages mapped", xlabel="Group", ylabel="RPKMs")
plt.xticks(rotation='90', rotation_mode='anchor', ha='right')
plt.savefig("../02_graphs/phages_rpkm_70_sample_barplot.svg")
plt.show()
plt.close()
```
